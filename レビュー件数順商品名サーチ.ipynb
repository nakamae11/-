```python
import requests

# あなたの楽天アプリID
APP_ID = "1016526773867927317"

# 検索キーワード（好きなものに変えてOK）
keyword = "コンセント収納"

# APIエンドポイントとパラメータ
url = "https://app.rakuten.co.jp/services/api/IchibaItem/Search/20170706"
params = {
    "applicationId": APP_ID,
    "keyword": "コンセント収納",
    "sort": "-reviewCount",  # 擬似的に「売れ筋商品順」
    "hits": 30,
    "format": "json"
}

# APIリクエスト送信
response = requests.get(url, params=params)
data = response.json()

# 結果を表示
for i, item in enumerate(data["Items"], 1):
    info = item["Item"]
    print(f"[{i}] 商品名：{info['itemName']}")
    print(f"    価格：{info['itemPrice']}円")
    print(f"    レビュー平均：{info['reviewAverage']}")
    print(f"    レビュー件数：{info['reviewCount']}")
    print(f"    商品URL：{info['itemUrl']}")
    print("-" * 50)
```
